```python
import sys
from pathlib import Path

sys.path.append(str(Path().resolve().parent))

from parsing.loader import load_candidates_jsonl
from features.experience import experience_score
from features.semantic import keyword_score
from features.behavior import behavior_score
from features.scorer import total_score
```

```python
candidates = load_candidates_jsonl(
    "../data/raw/candidates.jsonl",
    limit=1000
)

len(candidates)
```

```python
ranked = []

for candidate in candidates:
    ranked.append({
        "candidate": candidate,
        "experience": experience_score(candidate),
        "semantic": keyword_score(candidate),
        "behavior": behavior_score(candidate),
        "total": total_score(candidate)
    })

ranked.sort(
    key=lambda x: x["total"],
    reverse=True
)
```

```python
for rank, row in enumerate(ranked[:20], start=1):

    candidate = row["candidate"]

    print("=" * 100)

    print(f"Rank #{rank}")
    print(f"ID: {candidate.candidate_id}")

    print(
        f"{candidate.profile.current_title}"
        f" @ {candidate.profile.current_company}"
    )

    print()

    print(
        f"Experience Score : {row['experience']:.2f}"
    )

    print(
        f"Semantic Score   : {row['semantic']:.2f}"
    )

    print(
        f"Behavior Score   : {row['behavior']:.2f}"
    )

    print(
        f"TOTAL            : {row['total']:.2f}"
    )
```
